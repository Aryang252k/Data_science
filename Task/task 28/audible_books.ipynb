```python
def find_total_rating(txt):
     if pd.isna(txt):
        return txt
     try:
        ar=txt.split(' ')
        return int(ar[4].split('s')[2])  
    
     except Exception as e:
        return int(ar[4].split('s')[2].replace(',','')) 

def find_average_rating(txt):
     if pd.isna(txt):
        return txt
     try:
        ar=txt.split(' ')
        return float(ar[0])  
    
     except Exception as e:
        print(e)        
```

```python
df['total_rating']=df['stars'].apply(lambda x:find_total_rating(x))
df['average_rating']=df['stars'].apply(lambda x:find_average_rating(x))
```

```python
def to_min(tim):
   ar=tim.split()
   try:
     if ('hrs' in ar) and ('mins' in ar):
        return int(ar[0])*60+int(ar[3])
     elif 'hrs' in ar:
        return int(ar[0])*60
     elif 'mins' in ar:
        return int(ar[0])
   except:
        return 0
```

# Data Accessing

**About dataset**

* This data is of Iris flowers . It has an unusual structure and appearance. The petals resemble a classic fleur-de-lis symbol, with some petals rising up while others cascade down. The center petals stand upright and are known botanically as standards. These tall petals stand like signal flags, waving in potential pollinators. 

# Column's description 

* This dataset has six columns `Id`,`SepalLengthCm`,`SepalWidthCm`,`PetalLengthCm`,`PetalWidthCm`,`Species`.<br><br>
**Id**:- This column is useless.<br>
**SepalLengthCm**:- This column contain length of sepal i.e ("Sepals are small, leaf-shaped, green-coloured and outermost part of the flower") . The length of sepal is in centimeter.<br>
**SepalWidthCm**:- This column contain width of sepal in centimeter.<br>
**PetalLengthCm**:- This column contain length of petal i.e ("a sterile floral part that usually functions as a visually conspicuous element of a flower").The length of pental is in centimeter.<br>
**PetalWidthCm**:- This column contain width of petal in centimeter. <br>
**Species**:- This column contain species of iris flower i.e it is "Iris-setosa","Iris-versicolor" and "Iris-virginica".




```python
# Droping Id column as useless
data.drop(columns=['Id'],axis=1,inplace=True)
```

# Data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Iris_setosa=data[data['Species']=='Iris-setosa']
Iris_versicolor=data[data['Species']=='Iris-versicolor']
Iris_virginica=data[data['Species']=='Iris-virginica']
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(Iris_setosa.iloc[:,:-1],linewidths=0.5).set_title("Iris_setosa")
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(Iris_versicolor.iloc[:,:-1],linewidths=0.5).set_title("Iris_versicolor")
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(Iris_virginica.iloc[:,:-1],linewidths=0.5).set_title("Iris_virginica")
```

```python
plt.figure(figsize=(15,5))
sns.distplot(x=Iris_setosa.iloc[:,:-1],hist=False)
sns.distplot(x=Iris_versicolor.iloc[:,:-1],hist=False)
sns.distplot(x=Iris_virginica.iloc[:,:-1],hist=False)
plt.legend(labels=['Iris_setosa',"Iris_versicolor","Iris_virginica"])
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(Iris_setosa.iloc[:,:-1])
plt.legend(labels=['Iris_setosa',"Iris_versicolor","Iris_virginica"])
```

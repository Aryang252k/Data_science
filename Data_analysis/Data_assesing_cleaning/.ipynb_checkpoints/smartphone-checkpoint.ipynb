```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html,'lxml')
soup.prettify()
```

```python
phone=soup.find_all(class_='sm-product has-tag has-features has-actions')
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame()
for i in phone:
    name=[]
    price=[]
    rating=[]
    sim=[]
    processor=[]
    memory=[]
    battery=[]
    display=[]
    camera=[]
    card=[]
    os=[]
    
    name.append(i.find(class_='name clamp-2').text)
    price.append(i.find(class_='price').text)
    
    try:
        rating.append(i.find('b').text)
    except:
        rating.append(np.nan)
        
    try:
        sim.append(i.find_all('li')[0].text)
    except:
        sim.append(np.nan)
        
    try:
        processor.append(i.find_all('li')[1].text)
    except:
        processor.append(np.nan)
        
    try:
        memory.append(i.find_all('li')[2].text)
    except:
        memory.append(np.nan)
        
    try:
        battery.append(i.find_all('li')[3].text)
    except:
        battery.append(np.nan)
        
    try:
        display.append(i.find_all('li')[4].text)
    except:
        display.append(np.nan)
        
    try:
        camera.append(i.find_all('li')[5].text)
    except:
        camera.append(np.nan)
    
    try:
        card.append(i.find_all('li')[6].text)
    except:
        card.append(np.nan)
        
    try:
        os.append(i.find_all('li')[7].text)
    except:
        os.append(np.nan)
        
    data={'name':name,'price':price,'rating':rating,'sim':sim,'processor':processor,'memory':memory,'battery':battery,'display':display,'camera':camera,'card':card,'os':os}
    tempdf=pd.DataFrame(data)
    df=pd.concat([df,tempdf],ignore_index=True)
```

```python
df.tail(50)
```

```python
df.to_csv('smartphone.csv',index=False) # index=False to ignore index
```

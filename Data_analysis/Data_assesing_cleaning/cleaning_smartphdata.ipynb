# Summary of dataset
This data contains the infomation of smartphones with their specification like sim,processor,memory,battery,display type and size,card supported , operating system.

**`Column Description`**<br><br>
**name**-> This column contain name of smartphone.<br>
**price**-> This column contain price of smartphone.<br>
**rating**-> This column contain rating of for smartphone , the rating is out of 100. <br>
**sim**-> This column contain sim information i.e 3g/4g/5g connectivty and Dual Sim, 3G, 4G, 5G, VoLTE, Wi-Fi, NFC, IR Blaster.<br>
**processor**-> This column contain name of processor , number of core and its speed.<br>
**memory**-> This column contain size of RAM and also size of ROM. <br>
**batery**-> This column contain capacity of battery and type of charging i.e fast charging.<br>
**display**-> This column contain display size and its type and refresh rate.<br>
**camera**-> This column contain camera setup and megapixel.<br>
**card**-> This column contain memory card supported or not.<br>
**os**-> This column contain information of operating system of smartphone.
